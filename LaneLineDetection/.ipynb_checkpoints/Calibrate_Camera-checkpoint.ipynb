```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import cv2 
import pickle
import glob
```

## Calibrate Camera and solve calibration data in pickle file

```python
image_paths = glob.glob('adams_camera_cal/*.jpg')
#print('image paths: ',image_paths)
num_image_paths = len(image_paths)
print('number of image paths: ',num_image_paths)

nx = 9 # number of inside corners per row
ny = 6 # number of inside corners per col

"""
prepare object points:
[[0,0,0],
 [1,0,0],
 [2,0,0],
 ......,
 [8,5,0]
"""
objp = np.zeros((ny*nx,3),np.float32)
objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2)

# Lists to store object points and image points from all images
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image space

# Iterate through list of images and draw chessboard corners on image
for index,fileName  in enumerate(image_paths):
    image = cv2.imread(fileName)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (nx,ny), None)
    # If chessboard corners found
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        # Draw the corners
        cv2.drawChessboardCorners(image, (nx,ny), corners, ret)

ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints,
                                                   gray.shape[::-1], # tuple (width,height)
                                                   None, None)
print('mtx: ',mtx)
print('dist: ', dist)
file_name = 'calibration.p'

with open(file_name, 'wb') as f: 
    pickle.dump([mtx, dist], f)
```

## Load Pickle Data

```python
file_name = 'calibration.p'

with open(file_name, 'rb') as f:   
    mtx, dist = pickle.load(f)
    
print('mtx: ',mtx)
print('dist: ', dist)
```

## Checking to make sure camera is calibrated properly

```python
img = 'camera_cal/calibration1.jpg'
image = cv2.imread(img)
image_undistorted = cv2.undistort(image, mtx, dist, None, mtx)
f, axs = plt.subplots(1,2,figsize=(32,12))
axs = axs.ravel()
axs[0].imshow(image)
axs[0].set_title('original',fontsize=40)
axs[1].imshow(image_undistorted)
axs[1].set_title('undistorted',fontsize=40)
plt.show()
```
