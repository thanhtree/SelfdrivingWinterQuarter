## Helper Functions

```python
# Tuple Format: (Lane Line Number, Slope, y-Intercept)
def getLaneLineNumber(tupleList,intercept):
    for Tuple in tupleList:
        if intercept == Tuple[2]:
            #index = tupleList.index(Tuple)
            lane_line_number = Tuple[0]
            break   
    return lane_line_number

def avg(myList):
    return sum(myList)/len(myList)
    
```

## Image Preprocessing



```python
"""

Create viewing window, convert to gray scale, apply gaussian smoothing,
apply Canny Edge Detection,and apply mask to ignore pixels outside 
viewing window

"""

ysize = image.shape[0]
xsize = image.shape[1]
region_select = np.copy(image)

# Define a region of interest 
# (x=0, y=0) is in the upper left in image processing
imshape = image.shape
#imshape[0] == ysize
left_bottom = (50, imshape[0])
right_bottom = (900, ysize)
left_top = (480,300)
right_top = (500,300)

# Make copy of image to show where viewing window is
image_copy = np.copy(image)

# Draw Viewing Window
left_line = np.array([[left_bottom[0],left_bottom[1],left_top[0],left_top[1]]],dtype=np.int32)
top_line =  np.array([[left_top[0],left_top[1],right_top[0],right_top[1]]],dtype=np.int32)
right_line = np.array([[right_top[0],right_top[1],right_bottom[0],right_bottom[1]]],dtype=np.int32)
bottom_line = np.array([[right_bottom[0],right_bottom[1],left_bottom[0],left_bottom[1]]],dtype=np.int32)
viewing_window = [left_line,top_line,right_line,bottom_line]

for line in viewing_window:
    for x1,y1,x2,y2 in line:
        cv2.line(image_copy,(x1,y1),(x2,y2),(0,255,0),10)

# Convert to Grayscale
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

# Define a kernel size and apply Gaussian smoothing
kernel_size = 5
blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size),0)

# Define our parameters for Canny and apply
low_threshold = 50
high_threshold = 150
edges = cv2.Canny(blur_gray, low_threshold, high_threshold)

# Create a masked edges image

# Define blank image to start with
mask = np.zeros_like(edges)   
ignore_mask_color = 255
# vertice order: Bottom Left, Top Left, Top Right, Bottom Right
vertices = np.array([[left_bottom, left_top, right_top, right_bottom]], dtype=np.int32)
# Fill in pixels within viewing window
cv2.fillPoly(mask, vertices, ignore_mask_color)
# Returning image only where pixels are non-zero
masked_edges = cv2.bitwise_and(edges, mask)

fig, axs = plt.subplots(5,1, figsize=(12,24))
axs = axs.ravel()

axs[0].imshow(image_copy)
axs[0].set_title('Viewing Window')
axs[1].imshow(gray)
axs[1].set_title('Gray Scale')
axs[2].imshow(blur_gray)
axs[2].set_title('Gaussian Blur')
axs[3].imshow(edges)
axs[3].set_title('Canny Edge Detection')
axs[4].imshow(masked_edges)
axs[4].set_title('Masked Edges')
plt.show()
```

## Apply Hough Transform

```python
# Define the Hough transform parameters
# Make a blank the same size as our image to draw on
rho = 1 # distance resolution in pixels of the Hough grid
theta = np.pi/180 # angular resolution in radians of the Hough grid
threshold = 50     # minimum number of votes (intersections in Hough grid cell)
min_line_length = 1 #minimum number of pixels making up a line
max_line_gap = 15   # maximum gap in pixels between connectable line segments
line_image = np.copy(image)*0 # creating a blank to draw lines on

# Run Hough on edge detected image
# Output "lines" is an array containing endpoints of detected line segments
lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)

# Make copy of image to show where viewing window is
image_copy = np.copy(image)

# Draw lines from Hough Transform onto copy of original image
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(image_copy,(x1,y1),(x2,y2),(255,0,0),10)


plt.imshow(image_copy)
plt.title('Hough Transform')
plt.show()
```

## Extrapolating Lines onto Image

```python
def process_image(image):
    # NOTE: The output should be a color image (3 channel) for processing video below

    ysize = image.shape[0]
    xsize = image.shape[1]
    image_copy = np.copy(image)
    overlay = np.copy(image_copy)

    # (x=0, y=0) is in the upper left in image processing
    
    #imshape[0] == ysize
    minY = 315
    # Vertices for Viewing Window
    left_bottom = (50, ysize)
    right_bottom = (900, ysize)
    left_top = (460,minY)
    right_top = (500,minY)
    
    # Convert to Grayscale
    gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size),0)

    # Define our parameters for Canny and apply
    low_threshold = 50
    high_threshold = 150
    edges = cv2.Canny(blur_gray, low_threshold, high_threshold)

    # Create a masked edges image
    mask = np.zeros_like(edges)   
    ignore_mask_color = 255   
    # vertice order: Bottom Left, Top Left, Top Right, Bottom Right
    vertices = np.array([[left_bottom, left_top, right_top, right_bottom]], dtype=np.int32)
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_edges = cv2.bitwise_and(edges, mask)

    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 50     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 100 #minimum number of pixels making up a line
    max_line_gap = 100   # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on
    # threshold = 50 seems to be best
    # Run Hough on edge detected image
    # Output "lines" is an array containing endpoints of detected line segments
    lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                                min_line_length, max_line_gap)
    
    # Getting list of slopes and yIntercepts associated with each line
    slopes = []
    yIntercepts = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            # Get Slope
            slope = (y2 - y1)/(x2 - x1)
            # Get Slope in Degrees
            slopeDegrees = np.arctan(slope)*(180/np.pi)
            # Get yintercept
            intercept = y1 - slope*x1
            slopes.append(slopeDegrees)
            yIntercepts.append(intercept)

    # I need to classify each yintercept and slope with a lane line. Either lane line 1 or lane line 2.
    # I do this with a tuple: (Lane Line Number, slope, yintercept)
    
    # myTuplesList keeps of list of tuples, where each tuple describes one line given by the Hough Transform
    
    # In order to figure out which line corresponds to which lane line, I will classify the first line
    # from the Hough Transform as belonging to lane line 1 by default. To check if the next lines
    # correspond to lane line 1 or a different lane line, I look at the "error" aka the absolute difference
    # between the yintercepts between the current line and all other lines that have been classified in
    # myTuplesList. Whichever one has the smallest error, and assuming the error is smaller than the
    # errorThreshold of 50 pixels, then I will assume that it belongs to the lane line with the smallest error.
    # If the minimum error is greater than 50 pixels, then I will assume that it belongs to a new lane line that
    # has not yet been classified. 
    # 
    # errorDictionary maps each error to its corresponding yintercept so that I know which yintercept the
    # minimum error corresponds to. If the minimum error is smaller than 50 pixels, then I check the lane line
    # classification number associated with the y intercept and
    # classify the new line as being part of that lane line

    # listOfErrors is the list of all the errors which is reset every time I classify a new yintercept
    
    myTuplesList = []
    errorDictionary = {}
    listOfErrors = []
    # numLines is the number of lane lines I have identified
    # Ideally, numLines should always be 2. However, the code
    # below will handle cases where more than 2 lane lines
    # have been identified.
    numLines = 1;
    # If the difference in the yintercept is larger than the
    # errorThreshold for all the yintercepts I have already
    # classified, then I will assume that the current line
    # belongs to a new lane line that needs to be
    # classified.
    errorThreshold = 50
    for intercept in yIntercepts:
        index = yIntercepts.index(intercept)
        # Find the slope associated with the yintercept
        slope = slopes[index]
        listOfErrors = []
        # By default, classify the first line as belonging
        # to lane line 1
        if yIntercepts.index(intercept) == 0:
            myTuple = (1,slope,intercept)
            myTuplesList.append(myTuple)
        else:
            for Tuple in myTuplesList:
                error = abs(Tuple[2] - intercept)
                errorDictionary[error] = Tuple[2]
                listOfErrors.append(error)
            minError = min(listOfErrors)
            if minError < errorThreshold:
               interceptAssociatedWithMinError = errorDictionary[minError]
               lane_line_number = getLaneLineNumber(myTuplesList,interceptAssociatedWithMinError)
               myTuple = (lane_line_number,slope,intercept)
               myTuplesList.append(myTuple)
            else:
                numLines += 1
                #print('new numLines: ',numLines)
                myTuple = (numLines,slope,intercept)
                myTuplesList.append(myTuple)
            
    # Taking the average of all desired yintercepts and slopes and drawing it on image    
    #print('myTuplesList: ',myTuplesList)
    currentLine = 1
    #print('numLines: ',numLines)
    while currentLine <= numLines:
        tempSlopesList = []
        tempInterceptsList = []
        for Tuple in myTuplesList:
            if Tuple[0] == currentLine:
                # Only include lines that have absolute value slopes
                # between 20 and 45 degrees
                if abs(Tuple[1]) > 20 and abs(Tuple[1]) < 45:
                    # Convert slope from degrees back to
                    # pixels/pixels for drawing lane line onto image
                    tempSlopesList.append(Tuple[1])
                    tempInterceptsList.append(Tuple[2])
        if len(tempSlopesList) > 0:


            tempSlopeAvg = avg(tempSlopesList)
            tempInterceptAvg = avg(tempInterceptsList)
            # Convert slope back from degrees to rise/run format
            tempSlopeAvg = np.tan(tempSlopeAvg*np.pi/180)
            y1 = ysize
            x1 = (y1 - tempInterceptAvg)/tempSlopeAvg
            # If I do not add another 10 pixels of wiggle room,
            # then the two lane lines will cross each other and form a triangle
            y2 = minY + 10
            x2 = (y2 - tempInterceptAvg)/tempSlopeAvg
            ndarray = np.array([[x1,y1,x2,y2]],dtype=np.int32)
            #print(ndarray)
            for x1,y1,x2,y2 in ndarray:
                cv2.line(overlay,(x1,y1),(x2,y2),(255,0,0),10)
                opacity = 0.4
                cv2.addWeighted(overlay,opacity,image_copy,1-opacity,0,image_copy)
        else:
            print('No reasonable slopes found')
        currentLine += 1
        
    return image_copy
    
result = process_image(image)
result2 = process_image(image2)
result3 = process_image(image3)
result4 = process_image(image4)
result5 = process_image(image5)
result6 = process_image(image6)

#save_filename = './test_images_output/solidWhiteRight.jpg'
#cv2.imwrite(save_filename,result)

plt.imshow(result)
plt.title('Result 1')
plt.show()

plt.imshow(result2)
plt.title('Result 2')
plt.show()

plt.imshow(result3)
plt.title('Result 3')
plt.show()

plt.imshow(result4)
plt.title('Result 4')
plt.show()

plt.imshow(result5)
plt.title('Result 5')
plt.show()

plt.imshow(result6)
plt.title('Result 6')
plt.show()
```
